```python
fontsize = 10 #setting plot paremeters
width_in_cm = 14.
params = {'figure.figsize': (width_in_cm , width_in_cm  * 9./16.),
          'axes.labelsize': fontsize*1.5,
          'axes.titlesize': fontsize*2,
          'xtick.labelsize': fontsize,
          'ytick.labelsize': fontsize,
          'legend.fontsize': fontsize*1}
plt.rcParams.update(params)
```

```python
IR_MarketCurves=pd.read_excel("IRMarketCurve.xlsx",sheet_name = 1,index_col='Maturity (in years)')#Loading data
```

```python
IR = IR_MarketCurves.drop(columns=["Unnamed: 5", "Swap Pricing", "Unnamed: 7","Rate"])
IR
```

```python
data = IR.iloc[:,[1]]
dataa = IR.iloc[:,[2]]
```

```python
fig, ax = plt.subplots(1, 1)#plotting historical close prices
ax.plot(data, label ='ZC BOND VALUES')
ax.plot(dataa, label ='ZC BOND RATES')

ax.set_xlabel('Maturity (in years)',fontsize=10)
ax.tick_params(axis='x', rotation=40,)
ax.set_ylabel('Value',fontsize=10)
ax.legend()
plt.title("Evolution du ZC bond (in years)")
ax.grid(axis='x', linestyle='-')
ax.grid(axis='y', linestyle='-')
fig.tight_layout() 
plt.show()
```
