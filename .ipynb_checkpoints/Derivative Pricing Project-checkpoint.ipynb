```python
def ZC_rate_func(maturity):
   cs = CubicSpline(bonds_maturities,bonds_rates)
   return float(cs(maturity))

def ZC_price_func(maturity):
   cs = CubicSpline(bonds_maturities,bonds_prices)
   return float(cs(maturity))

instantaneous_forward_rates = -1 / bonds_prices * np.gradient(bonds_prices)

def f(maturity):
  cs = CubicSpline(bonds_maturities, instantaneous_forward_rates)
  return float(cs(maturity))
```

```python
ZC_rate_vec=[]
for m in (bonds_maturities):
  ZC_rate_vec.append(ZC_rate_func(m))

ZC_price_vec=[]
for m in (bonds_maturities):
  ZC_rate_vec.append(ZC_price_func(m))

#defining the Hull-White semi-closed formula as seen in class
def B(t,T,a):
  return (1-np.exp(-a*(T-t)))/a

def A(t,T,sigma,a):
  return ZC_price_func(T)/ZC_price_func(t) * np.exp(B(t,T,a) * f(t) - sigma**2 * (1-np.exp(-2*a*t) * B(t,T,a)**2)/(4*a))

def P(t,T,r,sigma,a):
  return A(t,T,sigma,a)*np.exp(-B(t,T,a)*r)
```

```python
#Comparison of ZC bond values with HW formula VS Market prices
sigma_test = 0.02
a_test = 0.1 
HW_ZC_prices = []
Mkt_ZC_prices = []
#Simple_ZC

for m in bonds_maturities:
  hw_zc = P(0,m,ZC_rate_func(m),sigma_test,a_test)
  HW_ZC_prices.append(hw_zc)

  mkt_zc = ZC_price_func(m)
  Mkt_ZC_prices.append(mkt_zc)

  #simple_zc = np.exp(-m*ZC_rate(m))
  #Simple_ZC.append(simple_zc)


plt.plot(bonds_maturities, HW_ZC_prices, 'r', label='Hull-White semi-closed estimate')
plt.plot(bonds_maturities,Mkt_ZC_prices, 'b', label="Market value")
#plt.plot(bonds_maturities,Simple_ZC, 'g-', label="Simple estimate")
plt.xlabel('Maturity (years)')
plt.ylabel('ZC bond value')
plt.legend()
plt.show()
```

```python
#calibrating parameters
def objective(x):
    sigma, a = x
    error = 0
    for m in bonds_maturities:
      error += (P(0,m,ZC_rate_func(m),sigma,a)-ZC_price_func(m))**2
    return error

#Initial guess values for alpha, sigma and their respective bounds
x0 = [0.01, 0.02] # initial guess
boundaries = [(0.01, 0.1), (0.01, 0.05)] # respective bounds for sigma and a

_min= minimize(objective, x0, bounds=boundaries)
sigma_cal, a_cal= _min.x

print (f"Optimal sigma :{sigma_cal:.5f}")
print(f"Optimal mean reversion coeff: {a_cal:.5f}")
```

```python
#applying optimal sigma and alpha to HW formula for bond prices

HW_ZC_prices = []
Mkt_ZC_prices = []


for m in bonds_maturities:
  hw_zc = P(0,m,ZC_rate_func(m),sigma_cal,a_cal)
  HW_ZC_prices.append(hw_zc)

  mkt_zc = ZC_price_func(m)
  Mkt_ZC_prices.append(mkt_zc)


plt.plot(bonds_maturities, HW_ZC_prices, 'r', label='Hull-White semi-closed estimate')
plt.plot(bonds_maturities,Mkt_ZC_prices, 'b', label="Market value")

plt.xlabel('Maturity (years)')
plt.ylabel('ZC bond value')
plt.legend()
plt.show()
#much better 
```

```python
def Caplet_price(N,t,T_start,K,sigma,a):
    
    k=1+K*0.25
    T_end=T_start+0.25# we suppose caplet pays 90 days after starting date ~ 0.25 years
    std = sigma * np.sqrt((1-np.exp(-2*a*(T_start-t)))/(2*a)) * B(T_start,T_end,a)
    h = 1/std * np.log(P(t,T_end,ZC_rate_func(T_end),sigma,a)/P(t,T_start,ZC_rate_func(T_start),sigma,a)*K) + std/2
    
    #Defining the put option price function
    ZBP = k * P(t,T_start,ZC_rate_func(T_start),sigma,a) * norm.cdf(-h + std) - P(t,T_end,ZC_rate_func(T_end),sigma,a) * norm.cdf(-h)
    
    #Using the relationship between a caplet and a put option price at t on a ZC between T_start and T_end 
    cpl_price=N*(k)*ZBP
    
    return cpl_price

```
